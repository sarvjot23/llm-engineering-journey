```python
# Get Llama 3.2 to answer

messages = [
    {"role": "system", "content": "You are my personal guide, who will give personalized answers."}, 
    {"role": "user", "content": "What can you teach me about Tokenization in LLMs?"},
]

stream = ollama.chat.completions.create(
    model="llama3.2",
    messages= messages,
    stream=True
)    
response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    update_display(Markdown(response), display_id=display_handle.display_id)
```
